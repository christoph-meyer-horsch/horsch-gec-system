```python
lines = []

#read lines
with open("../datasets/horsch-dataset/formatted.txt", "r", encoding="utf-8") as f:
    for line in f:
        lines.append(line.strip().replace("\n","").split(" "))
```

```python
abbrevs = ["Lt.", "lt.", "Stk.", "stk.", "wg.", "bzw."]
words = ["Laut", "laut", "Stück", "Stück", "wegen", "beziehungsweise"]
```

```python
for i in range(len(lines)):
    for j in range(len(lines[i]) - 1):
        if "." in lines[i][j]:
            if lines[i][j] in abbrevs:
                lines[i][j] = words[abbrevs.index(lines[i][j])]
            else:
                print(" ".join(lines[i]) + " | " + lines[i][j])
                inp = input()
                if inp == "":
                    lines[i][j] = lines[i][j]
                else:
                    if lines[i][j] not in abbrevs:
                        abbrevs.append(lines[i][j])
                        words.append(inp)
                    lines[i][j] = inp
```

```python
#read lines
with open("../datasets/horsch-dataset/formatted2.txt", "w", encoding="utf-8") as f:
    for item in lines:
        f.write("%s\n" % " ".join(item))
```
