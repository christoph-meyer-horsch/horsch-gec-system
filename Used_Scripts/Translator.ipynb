```python
import pycurl
from urllib.parse import urlencode
from io import BytesIO
import json
```

```python
lines = []
translations = []

#read lines
with open("../datasets/horsch-dataset/test_set.txt", "r", encoding="utf-8") as f:
    for line in f:
        lines.append(line.strip().replace("\n",""))
```

```python
url = "https://api.deepl.com/v2/translate"
auth_key = "2014225d-82e3-a8d7-050d-dfb43ac3a354"
target_lang = "EN"
```

```python
for line in lines:
    buffer = BytesIO()
    crl = pycurl.Curl()
    crl.setopt(crl.URL, url)
    data = {
        "auth_key" : auth_key,
        "text" : line,
        "target_lang" : target_lang
    }
    pf = urlencode(data)
    crl.setopt(crl.POSTFIELDS, pf)
    crl.setopt(crl.WRITEDATA, buffer)
    
    crl.perform()
    crl.close()
    
    body = buffer.getvalue().decode("UTF-8")
    print(body)
    data = json.loads(body)
    translations.append(data["translations"][0]["text"])
```

```python
with open("../datasets/horsch-dataset/test_set_eng.txt", "w") as f:
    f.write("\n".join(translations))
```
